## Combining features with Julian and Riya

```python
from datetime import datetime, timedelta
```

```python
counties = pd.read_csv("https://raw.githubusercontent.com/SammyAgrawal/BayAreaHax/master/County%20Red%3ABlue%20%25/Output/ma_county.csv")
counties.head()
```

```python
def dem_percent_county(county):
    return counties[counties['County']==county]['Democratic Percentage'].iloc[0]

def rep_percent(county):
    return counties[counties['County']==county]['Republican Percentage'].iloc[0]
```

```python
df["% Dem"] = df['County'].apply(dem_percent_county)
df['% Rep'] = df['County'].apply(rep_percent)
```

```python
demo = pd.read_excel('Demographics.xlsx')
demo.fillna(datetime.strptime('3/23/20','%m/%d/%y' ), inplace=True) # one blank entry, looked up value
```

```python
def weeksInOffice(date):
    if(isinstance(date, datetime)):
        delta = datetime.now() - date
        return(int(delta.days / 7))
    else: #int
        years = datetime.now().year - date
        return(int(years * 52))

demo["Weeks In Office"] = demo['Date Assumed Office'].apply(weeksInOffice)
demo[demo["Weeks In Office"]>0].count()
```

#### Adding Last name column to demoographic df- Names are too different, match using this

```python
last_names = [name.split(" ")[-1] for name in demo['Name']]
demo['Last'] = pd.Series(last_names)

demo['Last'][1] = 'Barrett' # III (1/2)
demo['Last'][69] = 'Schmid' #III (2/2)
demo['Last'][87] = 'González' #accent mark
demo['Last'][155] = "Donato" # Sr. (1/1)
for i in demo[demo['Last']=='Jr.'].index:
    real_last = demo.iloc[i]["Name"].split(" ")[1] #get middle word, not last
    demo['Last'][i] = real_last
```

```python
def getWiO(lname):
    # returns Weeks in Office from demographic dataframe- based on last name
    match = demo[demo['Last'] == lname]
    return match['Weeks In Office'].iloc[0]
```

```python
df['Weeks in Office'] = 0 # new column
for last_name in df['Last Name']:
    index = df[df['Last Name']==last_name].index[0]
    df['Weeks in Office'][index] = getWiO(last_name)
```

```python
def eNum(url):
    result = 0
    print('Start')
    keywords = ["Environment", 'environment', 'Natural', 'carbon'] #- sums up, so if both in same bill, counts twice
    response = requests.get(url, verify=False)
    page = Selector(text=response.text)
    bills = pd.Series(page.xpath("//tbody/tr/td[contains(@class, 'sponsoredBillTitle')]").getall())
    for kw in keywords:
        result += int(bills.astype(str).str.contains(kw).sum())
    return(result)
```

```python
url2 = 'https://malegislature.gov/Legislators/Profile/BJA1'
eNum(url2)
```

```python
df['No. Enviro'] = df['Link'].apply(eNum)
```

### Teachers and Programmers render themselves irrelevant

```python
def add_bill(name, url, year=191):
    bills[name] = 0 # new column
    resp = requests.get(url, verify=False)
    
    root = Selector(text=resp.text)
    petition_link = root.xpath('//tbody/tr/td/a/@href').getall()
    
    links = pd.Series(petition_link).apply(lambda p: (base_url + p.split('/'+str(year))[0]))
    for link in links:
        try:
            index = bills[bills['Link']==link].index[0]
            bills[name][index] = 1
        except:
            print('Not found: '+ link)  
```

```python
add_bill('H2007', 'https://malegislature.gov/Bills/191/H2007/Cosponsor')
```

```python
bills[bills['H2007']==1]
```
